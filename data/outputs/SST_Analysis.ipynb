```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("images/Curacao Coral Reef Assessment 2023 CUR/metadata.csv")

unique_dates = df["date"].unique()
dates_dic = {}

for date in unique_dates:
    dates_dic[date] = df[df["date"]==date].iloc[0]["SST@90th_HS"]

dates_dic = dict(sorted(dates_dic.items()))

clist = []
llist = []
for date in dates_dic:
    llist.append(df[df["SST@90th_HS"]==dates_dic[date]]["label"].mode()[0])
    if df[df["SST@90th_HS"]==dates_dic[date]]["label"].mode()[0]=="bleached":
        clist.append('r')
    else:
        clist.append('g')


plt.scatter(dates_dic.keys(), dates_dic.values(), marker='o', c=clist)
plt.plot(dates_dic.keys(), dates_dic.values())

plt.title("Time Series Progression of SST@90th_HS for Curacao Coral Reef Assessment 2023 CUR")
plt.xlabel("Date")
plt.ylabel("SST@90th_HS")
plt.xticks(rotation=45)
plt.show()
```
