```python
import matplotlib.pyplot as plt
```

```python
LR_001 = []
LR_005 = []

with open("ResNet_0.001LR_50Epochs.txt", "r") as f:
    for line in f.readlines():
        if "loss:" in line:
            LR_001.append(float(line.split(":")[1].strip()))

with open("ResNet_0.005LR_50Epochs.txt", "r") as f:
    for line in f.readlines():
        if "loss:" in line:
            LR_005.append(float(line.split(":")[1].strip()))
```

```python
x = [i+1 for i in range(50)]

plt.plot(x[:10], LR_001[:10], label="LR=0.001")
plt.plot(x[:10], LR_005[:10], label="LR=0.005")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("MSE Loss trend from epochs 1-10")
plt.legend()
plt.show()
```

```python
plt.plot(x[10:50], LR_001[10:50], label="LR=0.001")
plt.plot(x[10:50], LR_005[10:50], label="LR=0.005")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("MSE Loss trend from epochs 11-50")
plt.legend()
plt.show()
```
