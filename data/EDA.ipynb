```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the metadata files
ABB = pd.read_csv('outputs\images\Altieri Biscayne Bay\metadata.csv')
ARU = pd.read_csv('outputs\images\Curacao Coral Reef Assessment 2023 ARU\metadata.csv')
CUR = pd.read_csv('outputs\images\Curacao Coral Reef Assessment 2023 CUR\metadata.csv')
WAPACI = pd.read_csv('outputs\images\WAPA Coral Inventory 2.0\metadata.csv')
WAPACT = pd.read_csv('outputs\images\WAPA CoralNet Training\metadata.csv')
WAPAI = pd.read_csv('outputs\images\WAPA Interns\metadata.csv')

# Combine the metadata files
metadata = pd.concat([ABB, ARU, CUR, WAPACI, WAPACT, WAPAI])
metadata = metadata.reset_index(drop=True)
```

```python
# EDA on collect date
date = metadata['date']
dates = dict()
for d in reversed(date):
    if d in dates:
        dates[d] += 1
    else:
        dates[d] = 1
plt.figure(figsize=(14, 8))
plt.bar(dates.keys(), dates.values())
plt.xticks(rotation=45)
plt.tight_layout()
plt.title('Number of images collected per day')#3255 images in total
plt.show()

```

```python
# EDA on location
location = metadata['location']
locations = dict()
for l in location:
    if l in locations:
        locations[l] += 1
    else:
        locations[l] = 1
colors = ['blue', 'orange', 'green']
plt.pie(locations.values(), labels=None, colors=colors, autopct='%1.1f%%')
plt.legend(locations.keys(), title="Locations", loc="upper left", bbox_to_anchor=(0.9, 1))
plt.title('Distribution of images per location')

plt.show()
```

```python
#EDA on labels
label = metadata['label']
labels = dict()
for l in label:
    if l in labels:
        labels[l] += 1
    else:
        labels[l] = 1
color = ["pink", "purple", "grey"]
plt.pie(labels.values(), labels=None, colors=color, autopct='%1.1f%%')
plt.legend(labels.keys(), title="status", loc="upper left", bbox_to_anchor=(0.9, 1))
plt.title('Distribution of coral per status')
plt.show()
```
